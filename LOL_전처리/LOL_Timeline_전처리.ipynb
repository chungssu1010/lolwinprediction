```python
'''
'blue_baron_first',
'blue_champion_first',
'blue_dragon_first',
'blue_inhibitor_first',
'blue_riftherald_first',
'blue_tower_first',

'''
```

```python
# first 처리 함수

def first_kill(df, column_name, blue_column, red_column):
    blue_index = np.where(df[blue_column] >= 1)[0]
    red_index = np.where(df[red_column] >= 1)[0]
    df[f'blue_{column_name}_first'] = 0
    df[f'red_{column_name}_first'] = 0

    if blue_index.size > 0 and red_index.size > 0:
        if blue_index[0] < red_index[0]:
            df.loc[blue_index[0]:, f'blue_{column_name}_first'] = 1
        else:
            df.loc[red_index[0]:, f'red_{column_name}_first'] = 1
    elif blue_index.size > 0:
        df.loc[blue_index[0]:, f'blue_{column_name}_first'] = 1
    elif red_index.size > 0:
        df.loc[red_index[0]:, f'red_{column_name}_first'] = 1
```

```python
# baron_first
first_kill(df, 'baron', 'blue_baron', 'red_baron')

# champion_first
first_kill(df, 'champion', 'blue_champion', 'red_champion')

# inhibitor_first
first_kill(df, 'inhibitor', 'blue_inhibitor', 'red_inhibitor')

# dragon_first
first_kill(df, 'dragon', 'blue_dragon', 'red_dragon')

# riftherald_first
first_kill(df, 'riftherald', 'blue_riftherald', 'red_riftherald')

# tower_first
first_kill(df, 'tower', 'blue_tower', 'red_tower')
```

```python
import pandas as pd
import numpy as np
import ast

# first 처리 함수
def first_kill(df, column_name, blue_column, red_column):
    blue_index = np.where(df[blue_column] >= 1)[0]
    red_index = np.where(df[red_column] >= 1)[0]
    df[f'blue_{column_name}_first'] = 0
    df[f'red_{column_name}_first'] = 0

    if blue_index.size > 0 and red_index.size > 0:
        if blue_index[0] < red_index[0]:
            df.loc[blue_index[0]:, f'blue_{column_name}_first'] = 1
        else:
            df.loc[red_index[0]:, f'red_{column_name}_first'] = 1
    elif blue_index.size > 0:
        df.loc[blue_index[0]:, f'blue_{column_name}_first'] = 1
    elif red_index.size > 0:
        df.loc[red_index[0]:, f'red_{column_name}_first'] = 1

# 타임라인 전처리
def timeline_preprocessing(df, z):
    # df dict형태로 all_frames에 저장
    all_frames = []

    for i, row in df.iterrows():
        data = ast.literal_eval(row['info'])
        all_frames.append(data)

    # timestamp
    timestamp = []

    for i in range(len(all_frames[z]['frames'])):
        timestamp.append(i)

    # tower / inhibitor
    blue_tower, red_tower = [], []
    blue_inhibitor, red_inhibitor = [], []
    B_tower, R_tower = 0, 0
    B_inhibitor, R_inhibitor = 0, 0

    for i in range(len(all_frames[z]['frames'])):
        for j in range(len(all_frames[z]['frames'][i]['events'])):
            if all_frames[z]['frames'][i]['events'][j]['type'] == 'BUILDING_KILL':
                if all_frames[z]['frames'][i]['events'][j]['buildingType'] == 'INHIBITOR_BUILDING':
                    if all_frames[z]['frames'][i]['events'][j]['killerId'] == 0:
                        if all_frames[z]['frames'][i]['events'][j]['teamId'] == 100:
                            B_inhibitor += 1
                        else:
                            R_inhibitor += 1
                    else:
                        if all_frames[z]['frames'][i]['events'][j]['teamId'] == 100:
                            R_inhibitor += 1
                        else:
                            B_inhibitor += 1
                if all_frames[z]['frames'][i]['events'][j]['buildingType'] == 'TOWER_BUILDING':
                    if all_frames[z]['frames'][i]['events'][j]['killerId'] == 0:
                        if all_frames[z]['frames'][i]['events'][j]['teamId'] == 100:
                            B_tower += 1
                        else:
                            R_tower += 1
                    else:
                        if all_frames[z]['frames'][i]['events'][j]['teamId'] == 100:
                            R_tower += 1
                        else:
                            B_tower += 1

        blue_tower.append(B_tower)
        red_tower.append(R_tower)
        blue_inhibitor.append(B_inhibitor)
        red_inhibitor.append(R_inhibitor)
    # damage, level, minion, gold
    blue_damage, red_damage = [], []
    blue_level, red_level = [], []
    blue_minion, red_minion = [], []
    blue_gold, red_gold = [], []
    B_damage, R_damage = 0, 0
    B_level, R_level = 0, 0
    B_minion, R_minion = 0, 0
    B_gold, R_gold = 0, 0

    # damage
    for i in range(len(all_frames[z]['frames'])):
        damage_1 = all_frames[z]['frames'][i]['participantFrames']['1']['damageStats']['totalDamageDoneToChampions']
        damage_2 = all_frames[z]['frames'][i]['participantFrames']['2']['damageStats']['totalDamageDoneToChampions']
        damage_3 = all_frames[z]['frames'][i]['participantFrames']['3']['damageStats']['totalDamageDoneToChampions']
        damage_4 = all_frames[z]['frames'][i]['participantFrames']['4']['damageStats']['totalDamageDoneToChampions']
        damage_5 = all_frames[z]['frames'][i]['participantFrames']['5']['damageStats']['totalDamageDoneToChampions']
        damage_6 = all_frames[z]['frames'][i]['participantFrames']['6']['damageStats']['totalDamageDoneToChampions']
        damage_7 = all_frames[z]['frames'][i]['participantFrames']['7']['damageStats']['totalDamageDoneToChampions']
        damage_8 = all_frames[z]['frames'][i]['participantFrames']['8']['damageStats']['totalDamageDoneToChampions']
        damage_9 = all_frames[z]['frames'][i]['participantFrames']['9']['damageStats']['totalDamageDoneToChampions']
        damage_10 = all_frames[z]['frames'][i]['participantFrames']['10']['damageStats']['totalDamageDoneToChampions']
        B_damage = damage_1 + damage_2 + damage_3 + damage_4 + damage_5
        R_damage = damage_6 + damage_7 + damage_8 + damage_9 + damage_10
        blue_damage.append(B_damage/5)
        red_damage.append(R_damage/5)

    # level
    for i in range(len(all_frames[z]['frames'])):
        level_1 = all_frames[z]['frames'][i]['participantFrames']['1']['level']
        level_2 = all_frames[z]['frames'][i]['participantFrames']['2']['level']
        level_3 = all_frames[z]['frames'][i]['participantFrames']['3']['level']
        level_4 = all_frames[z]['frames'][i]['participantFrames']['4']['level']
        level_5 = all_frames[z]['frames'][i]['participantFrames']['5']['level']
        level_6 = all_frames[z]['frames'][i]['participantFrames']['6']['level']
        level_7 = all_frames[z]['frames'][i]['participantFrames']['7']['level']
        level_8 = all_frames[z]['frames'][i]['participantFrames']['8']['level']
        level_9 = all_frames[z]['frames'][i]['participantFrames']['9']['level']
        level_10 = all_frames[z]['frames'][i]['participantFrames']['10']['level']
        B_level = level_1 + level_2 + level_3 + level_4 + level_5
        R_level = level_6 + level_7 + level_8 + level_9 + level_10
        blue_level.append(B_level/5)
        red_level.append(R_level/5)

    # jungle minion
    blue_jungle_minion, red_jungle_minion = [], []
    B_jungle_minion, R_jungle_minion = 0, 0

    for i in range(len(all_frames[z]['frames'])):
        jungle_1 = all_frames[z]['frames'][i]['participantFrames']['1']['jungleMinionsKilled']
        jungle_2 = all_frames[z]['frames'][i]['participantFrames']['2']['jungleMinionsKilled']
        jungle_3 = all_frames[z]['frames'][i]['participantFrames']['3']['jungleMinionsKilled']
        jungle_4 = all_frames[z]['frames'][i]['participantFrames']['4']['jungleMinionsKilled']
        jungle_5 = all_frames[z]['frames'][i]['participantFrames']['5']['jungleMinionsKilled']
        jungle_6 = all_frames[z]['frames'][i]['participantFrames']['6']['jungleMinionsKilled']
        jungle_7 = all_frames[z]['frames'][i]['participantFrames']['7']['jungleMinionsKilled']
        jungle_8 = all_frames[z]['frames'][i]['participantFrames']['8']['jungleMinionsKilled']
        jungle_9 = all_frames[z]['frames'][i]['participantFrames']['9']['jungleMinionsKilled']
        jungle_10 = all_frames[z]['frames'][i]['participantFrames']['10']['jungleMinionsKilled']

    # minion
    blue_minion, red_minion = [], []
    B_minion, R_minion = 0, 0

    for i in range(len(all_frames[z]['frames'])):
        minion_1 = all_frames[z]['frames'][i]['participantFrames']['1']['minionsKilled']
        minion_2 = all_frames[z]['frames'][i]['participantFrames']['2']['minionsKilled']
        minion_3 = all_frames[z]['frames'][i]['participantFrames']['3']['minionsKilled']
        minion_4 = all_frames[z]['frames'][i]['participantFrames']['4']['minionsKilled']
        minion_5 = all_frames[z]['frames'][i]['participantFrames']['5']['minionsKilled']
        minion_6 = all_frames[z]['frames'][i]['participantFrames']['6']['minionsKilled']
        minion_7 = all_frames[z]['frames'][i]['participantFrames']['7']['minionsKilled']
        minion_8 = all_frames[z]['frames'][i]['participantFrames']['8']['minionsKilled']
        minion_9 = all_frames[z]['frames'][i]['participantFrames']['9']['minionsKilled']
        minion_10 = all_frames[z]['frames'][i]['participantFrames']['10']['minionsKilled']
        B_minion = minion_1 + minion_2 + minion_3 + minion_4 + minion_5 + jungle_1 + jungle_2 + jungle_3 + jungle_4 + jungle_5
        R_minion = minion_6 + minion_7 + minion_8 + minion_9 + minion_10 + jungle_6 + jungle_7 + jungle_8 + jungle_9 + jungle_10
        blue_minion.append(B_minion)
        red_minion.append(R_minion)

    # gold
    for i in range(len(all_frames[z]['frames'])):
        gold_1 = all_frames[z]['frames'][i]['participantFrames']['1']['totalGold']
        gold_2 = all_frames[z]['frames'][i]['participantFrames']['2']['totalGold']
        gold_3 = all_frames[z]['frames'][i]['participantFrames']['3']['totalGold']
        gold_4 = all_frames[z]['frames'][i]['participantFrames']['4']['totalGold']
        gold_5 = all_frames[z]['frames'][i]['participantFrames']['5']['totalGold']
        gold_6 = all_frames[z]['frames'][i]['participantFrames']['6']['totalGold']
        gold_7 = all_frames[z]['frames'][i]['participantFrames']['7']['totalGold']
        gold_8 = all_frames[z]['frames'][i]['participantFrames']['8']['totalGold']
        gold_9 = all_frames[z]['frames'][i]['participantFrames']['9']['totalGold']
        gold_10 = all_frames[z]['frames'][i]['participantFrames']['10']['totalGold']
        B_gold = gold_1 + gold_2 + gold_3 + gold_4 + gold_5
        R_gold = gold_6 + gold_7 + gold_8 + gold_9 + gold_10
        blue_gold.append(B_gold)
        red_gold.append(R_gold)

    # baron / dragon / riftherald
    blue_baron, red_baron = [], []
    blue_dragon, red_dragon = [], []
    blue_riftherald, red_riftherald = [], []
    B_baron, R_baron = 0, 0
    B_dragon, R_dragon = 0, 0
    B_riftherald, R_riftherald = 0, 0

    for i in range(len(all_frames[z]['frames'])):
        for j in range(len(all_frames[z]['frames'][i]['events'])):
            if all_frames[z]['frames'][i]['events'][j]['type'] == 'ELITE_MONSTER_KILL':
                if all_frames[z]['frames'][i]['events'][j]['monsterType'] == 'BARON_NASHOR':
                    if all_frames[z]['frames'][i]['events'][j]['killerTeamId'] == 100:
                        B_baron += 1
                    else:
                        R_baron += 1
                if all_frames[z]['frames'][i]['events'][j]['monsterType'] == 'DRAGON':
                    if all_frames[z]['frames'][i]['events'][j]['killerTeamId'] == 100:
                        B_dragon += 1
                    else:
                        R_dragon += 1
                if all_frames[z]['frames'][i]['events'][j]['monsterType'] == 'RIFTHERALD':
                    if all_frames[z]['frames'][i]['events'][j]['killerTeamId'] == 100:
                        B_riftherald += 1
                    else:
                        R_riftherald += 1

        blue_baron.append(B_baron)
        red_baron.append(R_baron)
        blue_dragon.append(B_dragon)
        red_dragon.append(R_dragon)
        blue_riftherald.append(B_riftherald)
        red_riftherald.append(R_riftherald)

    # kill / assist / death
    blue_champion, red_champion = [], []
    blue_death, red_death = [], []
    blue_assist, red_assist = [], []
    blue_solo_kill, red_solo_kill = [], []
    B_champion, R_champion = 0, 0
    B_death, R_death = 0, 0
    B_assist, R_assist = 0, 0
    B_solo_kill, R_solo_kill = 0, 0

    for i in range(len(all_frames[z]['frames'])):
        for j in range(len(all_frames[z]['frames'][i]['events'])):
            if all_frames[z]['frames'][i]['events'][j]['type'] == 'CHAMPION_KILL':
                if all_frames[z]['frames'][i]['events'][j]['killerId'] <= 5:
                    B_champion += 1
                    if 'assistingParticipantIds' in all_frames[z]['frames'][i]['events'][j]:
                        B_assist += len(all_frames[z]['frames'][i]['events'][j]['assistingParticipantIds'])
                    else:
                        B_solo_kill += 1
                elif all_frames[z]['frames'][i]['events'][j]['killerId'] > 5:
                    R_champion += 1
                    if 'assistingParticipantIds' in all_frames[z]['frames'][i]['events'][j]:
                        R_assist += len(all_frames[z]['frames'][i]['events'][j]['assistingParticipantIds'])
                    else:
                        R_solo_kill += 1
                if all_frames[z]['frames'][i]['events'][j]['victimId'] <= 5:
                    B_death += 1
                elif all_frames[z]['frames'][i]['events'][j]['victimId'] > 5:
                    R_death += 1

        blue_champion.append(B_champion)
        red_champion.append(R_champion)
        blue_death.append(B_death)
        red_death.append(R_death)
        blue_assist.append(B_assist)
        red_assist.append(R_assist)
        blue_solo_kill.append(B_solo_kill)
        red_solo_kill.append(R_solo_kill)

    # timeline 데이터 프레임
    timeline = pd.DataFrame({'timestamp' : timestamp, 'blue_tower' : blue_tower, 'red_tower' : red_tower,
              'blue_inhibitor' : blue_inhibitor, 'red_inhibitor' : red_inhibitor, 'blue_damage' : blue_damage,
              'red_damage' : red_damage, 'blue_level' : blue_level, 'red_level' : red_level,
              'blue_minion' : blue_minion, 'red_minion' : red_minion, 'blue_gold' : blue_gold, 'red_gold' : red_gold,
              'blue_baron' : blue_baron, 'red_baron' : red_baron, 'blue_dragon' : blue_dragon, 'red_dragon' : red_dragon,
              'blue_riftherald' : blue_riftherald, 'red_riftherald' : red_riftherald, 'blue_champion' : blue_champion,
              'red_champion' : red_champion, 'blue_death' : blue_death, 'red_death' : red_death, 'blue_assist' : blue_assist,
              'red_assist' : red_assist, 'blue_solo_kill' : blue_solo_kill, 'red_solo_kill' : red_solo_kill})

    # kda
    timeline['blue_kda'] = round((timeline['blue_champion'] + timeline['blue_assist']) / timeline['blue_death'], 2)
    timeline['red_kda'] = round((timeline['red_champion'] + timeline['red_assist']) / timeline['red_death'], 2)

    # kda NaN 처리
    timeline['blue_kda'] = timeline['blue_kda'].fillna(0)
    timeline['red_kda'] = timeline['red_kda'].fillna(0)

    # kda inf 처리
    timeline['blue_kda'] = np.where(timeline['blue_kda'] == np.inf, timeline['blue_champion'] + timeline['blue_assist'], timeline['blue_kda'])
    timeline['red_kda'] = np.where(timeline['red_kda'] == np.inf, timeline['red_champion'] + timeline['red_assist'], timeline['red_kda'])

    # gold_perM
    timeline['blue_gold_perM'] = round(timeline['blue_gold'] / timeline['timestamp'], 2)
    timeline['red_gold_perM'] = round(timeline['red_gold'] / timeline['timestamp'], 2)

    # gole_perM 0, 1분 데이터 0으로 변경
    timeline.at[0, 'blue_gold_perM'] = 0
    timeline.at[1, 'blue_gold_perM'] = 0
    timeline.at[0, 'red_gold_perM'] = 0
    timeline.at[1, 'red_gold_perM'] = 0

    # baron_first
    first_kill(timeline, 'baron', 'blue_baron', 'red_baron')

    # champion_first
    first_kill(timeline, 'champion', 'blue_champion', 'red_champion')

    # inhibitor_first
    first_kill(timeline, 'inhibitor', 'blue_inhibitor', 'red_inhibitor')

    # dragon_first
    first_kill(timeline, 'dragon', 'blue_dragon', 'red_dragon')

    # riftherald_first
    first_kill(timeline, 'riftherald', 'blue_riftherald', 'red_riftherald')

    # tower_first
    first_kill(timeline, 'tower', 'blue_tower', 'red_tower')

    # win
    timeline['blue_win'] = 0
    timeline['red_win'] = 0

    len_frames = len(all_frames[z]['frames'])
    len_events = len(all_frames[z]['frames'][len_frames - 1]['events'])

    win_team = all_frames[z]['frames'][len_frames - 1]['events'][len_events - 1]['winningTeam']
    if win_team == 100:
        timeline['blue_win'] = 1
    else:
        timeline['red_win'] = 1

    return timeline
```
