## PUUID -> MATCH ID

```python
# 그마
import pandas as pd
import requests
import time

# PUUID 데이터가 저장된 CSV 파일 경로
puuid_csv_file_path = '그마_puuid.csv'

# PUUID 데이터를 DataFrame으로 읽어오기
puuid_df2 = pd.read_csv(puuid_csv_file_path)

# 각 소환사의 PUUID를 사용하여 매치 ID 조회
match_ids_list = []

# 초당 및 2분당 요청 제한 설정
requests_per_second = 20  # 초당 요청 수
requests_per_two_minutes = 100  # 2분에 100개의 요청
two_minutes_counter = 0
two_minutes_limit = requests_per_two_minutes

for puuid in puuid_df2['0']:
    try:
        match_ids_url = f"https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid}/ids?start=0&count=100"
        r = requests.get(match_ids_url, headers=request_header)
        match_ids = r.json()
        match_ids_list.append(match_ids)
    except Exception as e:
        print(f"PUUID '{puuid}'의 매치 ID 조회 중 오류 발생: {str(e)}")

  # 1초의 대기 시간 설정
    time.sleep(1)

    # 초당 요청 카운터 업데이트
    two_minutes_counter += 1

    # 2분에 100개의 요청 제한을 초과하면 대기
    if two_minutes_counter >= two_minutes_limit:
        print("2분 대기 중...")
        time.sleep(120)  # 2분 대기
        two_minutes_counter = 0  # 카운터 초기화

# 각 소환사의 매치 ID 리스트를 저장
puuid_df2['matchIds'] = match_ids_list

# 매치 ID 정보가 추가된 DataFrame을 CSV 파일로 저장
puuid_df2.to_csv('puuid_data2_with_match_ids2.csv', index=False)

print("작업완료")
```

```python
import pandas as pd
import requests
import time

# 각 소환사의 PUUID를 사용하여 매치 ID 조회
match_history_data = []

# 초당 및 2분당 요청 제한 설정
requests_per_second = 20  # 초당 요청 수
requests_per_two_minutes = 100  # 2분에 100개의 요청
two_minutes_limit = requests_per_two_minutes

# 요청 제한을 고려하여 요청 시간 간격 계산
def calculate_request_interval(requests_per_second, requests_per_two_minutes):
    interval = 1.0 / requests_per_second
    if requests_per_two_minutes > 0:
        interval_2min = 120 / requests_per_two_minutes
        return max(interval, interval_2min)
    return interval

request_interval = calculate_request_interval(requests_per_second, requests_per_two_minutes)

for matchid in total_list['0'][21001:22000]:
    try:
        match_history = f"https://asia.api.riotgames.com/lol/match/v5/matches/{matchid}"
        r = requests.get(match_history, headers=request_header)
        match_data = r.json()
        match_history_data.append(match_data)
    except Exception as e:
        print(f"매치 ID '{matchid}' 조회 중 오류 발생: {str(e)}")

    # 요청 간격만큼 대기
    time.sleep(request_interval)

# 각 소환사의 매치 ID 리스트를 저장
match_result = pd.DataFrame(match_history_data)

# 매치 ID 정보가 추가된 DataFrame을 CSV 파일로 저장
match_result.to_csv('match_result_21001_22000.csv', index=False)

print("작업 완료")
```
